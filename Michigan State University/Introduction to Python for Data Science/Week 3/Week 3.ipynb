```python
import pandas as pd
import numpy as np


import warnings 
warnings.filterwarnings('ignore')
```

```python
# ENERGY
Energy = pd.read_excel('Energy Indicators.xls')
header_drop = list(range(0,17))
Energy.drop(header_drop, inplace = True)
footer_drop = list(range(244,282))
Energy.drop(footer_drop, inplace = True)
Energy.drop(['Unnamed: 0', 'Unnamed: 1'], axis = 1, inplace = True)
translate_cols = {
        'Unnamed: 2' : 'Country' ,
        'Unnamed: 3' : 'Energy Supply' ,
        'Unnamed: 4' : 'Energy Supply per Capita',
        'Unnamed: 5' : '% Renewable'
    }
Energy.rename(columns = translate_cols, inplace = True)  
Energy['Country'].replace('( \(.*\))','', regex=True, inplace = True)
Energy['Country'].replace('(\d+)','', regex=True, inplace = True)   
Energy['Country'] = Energy['Country'].str.strip()    
for col in Energy:
    Energy[col] = Energy[col].replace('...',np.nan)       
Energy['Energy Supply'] = Energy['Energy Supply'] * 1000000   
Energy['Country'] = Energy['Country'].str.replace('Republic of Korea','South Korea')
Energy['Country'] = Energy['Country'].str.replace('United States of America','United States')
Energy['Country'] = Energy['Country'].str.replace('United Kingdom of Great Britain and Northern Ireland','United Kingdom')
Energy['Country'] = Energy['Country'].str.replace('China, Hong Kong Special Administrative Region','Hong Kong')
    

#GDP
GDP = pd.read_csv('world_bank.csv', header = 3)
GDP.columns = GDP.iloc[0]
GDP.drop(0, inplace = True)       
GDP.rename(columns = {'Country Name' : 'Country'}, inplace = True)
GDP['Country'] = GDP['Country'].str.replace('Korea, Rep.','South Korea')
GDP['Country'] = GDP['Country'].str.replace('Iran, Islamic Rep.','Iran')
GDP['Country'] = GDP['Country'].str.replace('Hong Kong SAR, China','Hong Kong')    
GDP = GDP.drop(GDP.iloc[:,1:50], axis=1)    
GDP.columns = GDP.columns.astype(str).str.split('.').str[0]



#ScimEn
ScimEn = pd.read_excel('scimagojr-3.xlsx')



#Inner Join
merge1 = pd.merge(Energy, GDP, 'inner', 'Country')
merge2 = pd.merge(merge1, ScimEn, 'inner', 'Country')
df = merge2
df = df.sort_values(by = ['Rank'])
df.set_index('Country', inplace = True)
cols1 = ['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', '% Renewable', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015' ]
df = df[cols1]

#TOP 15
df = df.head(15)
    
```
