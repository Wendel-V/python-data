```python
def answer_eleven():
    df = answer_one()
    df['Population'] = df['Energy Supply'] / df['Energy Supply per Capita']
    df = df['Population']

    ContinentDict  = {'China':'Asia', 
                      'United States':'North America', 
                      'Japan':'Asia', 
                      'United Kingdom':'Europe', 
                      'Russian Federation':'Europe', 
                      'Canada':'North America', 
                      'Germany':'Europe', 
                      'India':'Asia',
                      'France':'Europe', 
                      'South Korea':'Asia', 
                      'Italy':'Europe', 
                      'Spain':'Europe', 
                      'Iran':'Asia',
                      'Australia':'Australia', 
                      'Brazil':'South America'}

    size = df.groupby(ContinentDict).size()
    sum = df.groupby(ContinentDict).sum()
    mean = df.groupby(ContinentDict).mean()
    std = df.groupby(ContinentDict).std()



    size = size.rename('size')
    sum = sum.rename('sum')
    mean = mean.rename('mean')
    std = std.rename('std')

    df = pd.concat([size, sum, mean, std], axis = 1)

    return df
```

```python
def anwer_twelve():
    df = answer_one()
    ContinentDict  = {'China':'Asia', 
                          'United States':'North America', 
                          'Japan':'Asia', 
                          'United Kingdom':'Europe', 
                          'Russian Federation':'Europe', 
                          'Canada':'North America', 
                          'Germany':'Europe', 
                          'India':'Asia',
                          'France':'Europe', 
                          'South Korea':'Asia', 
                          'Italy':'Europe', 
                          'Spain':'Europe', 
                          'Iran':'Asia',
                          'Australia':'Australia', 
                          'Brazil':'South America'}

    df['Continent'] = df.index.to_series().map(ContinentDict)

    df['Bins'] = pd.cut(df['% Renewable'], 5)

    df = df[['Continent', 'Bins']]

    return df.groupby(['Continent', 'Bins']).size()
```

```python
def answer_thirteen():
    df = answer_one()
    df = df[['Energy Supply', 'Energy Supply per Capita']]
    df['Population'] = df['Energy Supply'] / df['Energy Supply per Capita']
    df = df['Population']
    convert_pop = []
    for item in df:
        a, b = str(item).split(sep='.')
        a = a[::-1]
        new_number = []
        count = 1
        for digit in a:
            if count != 3:
                new_number.append(digit)
                count += 1
            else:
                new_number.append(digit)
                new_number.append(',')
                count = 1

        if new_number[-1] == ',':
            new_number.pop(-1)        
        new_number  = ''.join(new_number[::-1])    
        convert_pop.append(new_number+'.'+b)

    PopEst = pd.Series(convert_pop, index = df.index)
    return PopEst
```
